```python
import os
import numpy as np
```

```python
categories = os.listdir("./BBC News Summary/News Articles")[1:]
files = []
for category in categories:
    files.extend(
        os.path.join("./BBC News Summary/News Articles", category, file)
        for file in os.listdir("./BBC News Summary/News Articles/" + category)
    )

files, len(files)
```

```python
g = np.random.Generator(np.random.PCG64(42))
random_sample = g.choice(files, 1000, replace=False)
```

```python
data = ""

for f in random_sample:
    with open(f, "r", encoding='utf-8-sig', errors='ignore') as file:
        data += file.read()

with open("random_news_data.txt", "w", encoding='utf-8-sig') as file:
    file.write(data)
```
